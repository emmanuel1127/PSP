```python
# Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the excel file from directory after uploading to google collab

df=pd.read_excel('/Group Assessment 1.xlsx')
```

```python
# To view uploaded file
df.head()
```

```python
#To identify duplicates
df.duplicated()
```

```python
#To remove duplicates and update dataframe use (inplace=True)
df.drop_duplicates(inplace=True)
```

```python
# Perform the replacement of M to Married and S to Single in column 2, (Marital Status)

df['Marital Status'] = df['Marital Status'].str.replace('M', 'Married')
df['Marital Status'] = df['Marital Status'].str.replace('S', 'Single')
```

```python
#To view change/manipulation in column 2
print (df)
```

```python
# Perform the replacement of M to Male and F to Female in column 3, (Gender)

df['Gender'] = df['Gender'].str.replace('M', 'Male')
df['Gender'] = df['Gender'].str.replace('F', 'Female')
```

```python
#To view change/manipulation in column 3
print (df)
```

```python
# To round the income column to 0 decimal places changing them to integers.
#Note the income column was already displayed as integers.

df['Income'] = df['Income'].round(0)
```

```python
#Displaying the datatype of the income column
print (df['Income'].dtype)
```

```python
#To view updated dataframe
print (df)
```

```python
#To create a new column, 'Age Bracket' with conditionals.

# Define conditions and age_bracket

conditions = [
    (df['Age'] <= 30),
    (df['Age'] >= 31) & (df['Age'] <= 54),
    (df['Age'] > 55)
]
age_bracket = ['Adolescence', 'Middle-age', 'Old']

# To create the new column
df['Age Bracket'] = np.select(conditions, age_bracket, default='Unknown')
```

```python
#To view the column Age Bracket
print (df['Age Bracket'])
```

```python
# Perform the replacement of 10+ to More than 10 in column 10, (Commute Distance)

df['Commute Distance'] = df['Commute Distance'].str.replace('10+', 'More than 10')
```

```python
#To view change/manipulation in column 10
print (df['Commute Distance'])
```

```python
#Pivot tables and charts.
```

```python
#a. The average income of each gender and their bike purchase history

pivot_table = df.pivot_table(values='Income', index='Gender', columns='Purchased Bike', aggfunc='mean')
print(pivot_table)
```

```python
# Plot the pivot table data as a bar chart
pivot_table.plot(kind='bar', color=['blue', 'red'])

# Customize the plot
plt.title('Gender and Bike Purchase History')
plt.xlabel('Gender')
plt.ylabel('Average Income')
plt.xticks(rotation=45)
plt.legend(title='Purchased Bike')
plt.show()
```
